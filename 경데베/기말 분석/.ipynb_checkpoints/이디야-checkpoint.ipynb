## 이디야 ÷ 스타벅스
* 각 도시 별로 스타벅스 매장 1개에 몇개의 이디야가 있는지 확인

### 데이터 다듬기

* 앞에 있던 스타벅스 매장 수 데이터들을 테이블 형식으로 바꾸는 과정

```python
sb4 = starbucks1.City.value_counts().to_frame()
sb4.head()
```

```python
sb4["Number_S"] = sb4["City"]
sb4.head()
```

```python
sb4["City"] = sb4.index
sb4.head()
```

* 이디야 매장 수 데이터도 테이블 형식으로 다듬기

```python
ed4 = ediya1.City.value_counts().to_frame()
ed4.head()
```

```python
ed4["Number_E"] = ed4["City"]
ed4["City"] = ed4.index
ed4.head()
```

### 다듬은 데이터들을 합치기

* City를 중심으로 merge함수 이용
* Number_S는 스타벅스 매장 수
* Number_E는 이디야 매장 수

```python
sbed4 = pd.merge(sb4,ed4, on="City")
sbed4
```

* 병합한 데이터들을 따로 엑셀로 저장하기

```python
sbed4.to_csv("sb_ed.csv", mode='w')
```

### 이디야매장수 / 스타벅스 매장수
* E/S 열 생성

```python
sbed4["E/S"] = (sbed4["Number_E"] / sbed4["Number_S"])
sbed4
```

### 기존 데이터에 인구수도 추가
* 다시 merge를 이용해서 pop 데이터와 병합

```python
sbed_o2 = pd.merge(sbed4, pop, on="City")
sbed_o2
```

* 따로 저장

```python
sbed_o2.to_csv("sbedo.csv", mode='w')
```

## 상관관계 확인

### 인구수와 E/S비율간의 상관관계 확인

* 전체 도시 확인

```python
np.corrcoef(sbed9["E/S"], sbed9["Population"])
```

⇒ 음의 상관관계이지만 크게 관련은 없는것으로 나옴

### 상관관계를 그래프로 그려서 쉽게 확인해보자

```python
graph2 = np.polyfit(sbed9['Population'], sbed9['E/S'], 1)

g1 = np.poly1d(graph2) # y축
gx = np.linspace(100000, 700000, 100) # x축

# 오차를 추가한다.
sbed9['Mi'] = np.abs(5 - g1(sbed9['Population']))

gp_sort = sbed9.sort_values(by='Mi', ascending=False)
gp_sort.head()
```

```python
plt.figure(figsize=(20,14))
plt.scatter(sbed9['Population'], sbed9['E/S'], 
            c=sbed9['Mi'], s=50)
plt.plot(gx, g1(gx), ls='dashed', lw=3, color='r')

for n in range(10):
    plt.text(sbed9['Population'][n]*1.02, sbed9['E/S'][n]*0.98, 
             sbed9.index[n], fontsize=15)
    
plt.xlabel('Population')
plt.ylabel('E/S')
plt.colorbar()
plt.grid()
plt.show()
```

⇒ 확인이 어렵다

### 인구수가 적은 도시만 확인
* 전체를 한번에 확인하기에는 무리가 있어보인다 
* 그렇다면 인구수가 적은 도시 20개만 이용하여 따로 확인해보자

```python
graph_t = np.polyfit(sbed_ttt['Population'], sbed_ttt['E/S'], 1)

gt1 = np.poly1d(graph_t) # y축
gtx = np.linspace(100000, 700000, 100) # x축

# 오차를 추가한다.
sbed_ttt['Mi'] = np.abs(5 - gt1(sbed_ttt['Population']))

gpt_sort = sbed_ttt.sort_values(by='Mi', ascending=False)
gpt_sort.head()
```

```python
plt.figure(figsize=(15,15))
plt.scatter(sbed_ttt['Population'], sbed_ttt['E/S'], 
            c=sbed_ttt['Mi'], s=50)
plt.plot(gtx, gt1(gtx), ls='dashed', lw=3, color='r')

for n in range(10):
    plt.text(sbed_ttt['Population'][n]*1.02, sbed_ttt['E/S'][n]*0.98, 
             sbed_ttt.index[n], fontsize=15)
    
plt.xlabel('Population')
plt.ylabel('E/S')
plt.colorbar()
plt.grid()
plt.show()
```

## 상관분석, 그래프를 이용하여 E/S와 인구수의 상관관계를 알아보았다

### 분석을 해보니 인구수가 많은 도시 낮은 음의 상관관계를

### 인구수가 적은 도시는 비교적 뚜렷한 양의 상관관계를 나타낸다는 걸 알 수 있었다

### 왜 도시의 인구 수 별로 상관관계에 차이가 있는 것일까?

* 전체 매장수는 이디야가 700개 이상 많지만 원형그래프에서 볼 수 있듯 스타벅스는 몇개 도시에 매장수가 몰려있음

* 만약 속설처럼 이디야가 스타벅스 매장옆을 따라간다면 모든 도시에서 비슷한 상관관계를 가졌을것이다

* 그렇지만 실제 상관관계는 그렇지 않았다 왜냐하면 스타벅스는 대도시 내 매장 비중이 굉장히 높았고

* 반면 이디야는 인구수가 적은 중소도시에도 매장들을 많이 가지고 있다

## ⇒결론1

### 이디야가 스타벅스 옆에 존재한다는 속설은 결국 절반만 맞는 말이라고 할 수 있다

확대해서 지도를 본다면 스타벅스 주변에 이디야가 입점해 있다는 것을 확인 할 수있다

하지만 지도 전체를 본다면 스타벅스가 없는 중소도시에 이디야가 많이 존재한다는 것을 알 수 있다

또한 이디야가 스타벅스 주변에만 입점하는 전략을 사용하였다면 상관관계를 확인 하였을 때,

인구 수와 상관없이 서로 비슷한 양의 상관관계를 나타냈어야한다

하지만 전체의 도시를 따졌을 때 상관관계가 거의 없는 것으로 나왔고, 인구 수 상위 20개 도시는 오히려 음의 상관관계가 나왔다

이를 통해, 우리는 이디야가 스타벅스 주변 상권에 입점하는 전략만을 사용하는 것이 아니란 것을 알 수 있다

## 이디야는 이미 스타벅스와의 차별화를 위해 매장을 중소도시에 입점을 시키고 있다
## 만약 새로운 매장을 입점시키려고 한다면 어느 도시가 가장 유리할지 알아보자

먼저 인구수가 많아야 고객도 많을테니 인구수가 30만명 이상인 곳으로 제한한다

스타벅스가 너무 많아도 이디야가 너무 많아도 안된다

따라서 일단 E/S가 1.5를 넘지 않게 한다(이디야의 매장이 너무 많은 곳을 제외)

* 데이터 불러오기

```python
sbed000 = pd.read_csv('sbedo.csv',
                       usecols = ['City', 'Number_S', 'Number_E', 'E/S', 'Population'])
sbed000.head()
```

* 인구수 30만명 이상인 도시로 제한

```python
sbed001 = sbed000[sbed000.Population > 300000]
sbed001
```

* 인구수 순서로 정렬

```python
sbed001.sort_values(by="Population", ascending = False)
```

* 이디야가 너무 많은 곳 제외 ( E/S < 1.5)

```python
sbed002 = sbed001[sbed001.E/S < 1.5]
sbed002
```

⇒ / 때문에 읽혀지지가 않으니 이름을 바꾸자

```python
sbed001.rename(columns={sbed001.columns[3]: 'ES'}, inplace=True)
sbed001.head()
```

* 이름을 바꿨으니 다시 시도

```python
sbed002 = sbed001[sbed001.ES < 1.5]
sbed002
```

## ⇒결론2

### 인구수가 30만명 이상인 도시이면서 이디야 매장수가 스타벅스보다 이미 1.5배 이상 많지 않은 도시는 11개 뿐이다

* 만약 스타벅스 주변 상권에 입점하는 전략을 취한다면

    ⇒ 대구, 대전, 광주, 성남, 제주가 가장 적합할 것이다

* 만약 스타벅스가 없는 곳으로 입점을 하려고 한다면 인구수 대비 스타벅스가 매장 수가 적은

    ⇒대구, 제주, 천안, 포항, 구미 정도가 적합해 보인다

### 따라서 새롭게 매장을 입점 시켜야 한다면
### 대구 (스타벅스 주변에), 천안 (스타벅스가 없는 곳에)
### 이 두 도시에 입점을 시켜야한다
